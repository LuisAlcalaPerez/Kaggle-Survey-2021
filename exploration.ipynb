## Loading data and initial analysis

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pycountry as pyc
from graphics import *
```

```python
data = pd.read_csv('kaggle_survey_2021_responses.csv')
print(f'Rows: {data.shape[0]} \nCols: {data.shape[1]}')
data.head(10)
```

```python
# We can see that the first row is a description of each of the columns, let's separate that
descriptions = data.iloc[0,:]  # descriptions of the questions
data = data.iloc[1:,:]         # rest of the data

descriptions
```

```python
# We check now how many people participated in the survey, and how many of them are from Spain

n_participants = data.shape[0]                                                                                                             # total number of participants
n_questions = data.shape[1]                                                                                                                # total number of questions

data_spain = data[data.Q3 == 'Spain']                                                                                                      # dataframe only containning participants from Spain
n_participants_spain = data_spain.shape[0]                                                                                                 # number of participants from Spain

# Print the results
print(f'Number of participants: {n_participants}')
print(f'Number of participants from Spain: {n_participants_spain}, {((n_participants_spain*100)/n_participants):.2f}% of the total')
print(f'Number of questions: {n_questions}')
```

```python
# Lets see how many countries participated in the survey
list_countries = sorted(data[data.columns[3]].unique())                                       # list with all the countries that participated
print(f'Total number of countries that participated in the survey: {len(list_countries)}')
list_countries
```

```python
# As we can see from the countries list, there are some countries which names are no 'standard'
# We are going to make use of pycountry to rename those countries
list_official_countries = []
not_official_countries = []

for country in pyc.countries:                           # get a list of the official countries
    list_official_countries.append(country.name)

for item in list_countries:                             # get a list of the countries of the surveyt not appearing in the official list
    if item not in list_official_countries:
        not_official_countries.append(item)

not_official_countries
```

```python
# we replace the ones that we estimate
list_to_remove = ['Czech Republic', 'Russia', 'South Korea', 'Taiwan', 'United States of America', 'I do not wish to disclose my location', 'Other']
not_official_countries = [country for country in not_official_countries if country not in list_to_remove]
print(not_official_countries)

data[data.columns[3]].replace(
    {
        'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom',
        'Iran, Islamic Republic of...': 'Iran',
        'Hong Kong (S.A.R)': 'Hong Kong'
    }, inplace = True
)
```
